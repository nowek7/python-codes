```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
train = pd.read_pickle('../data/train.p')
test = pd.read_pickle('../data/test.p')
signnames = pd.read_csv('../data/signnames.csv')
```

```python
xTrain, yTrain = train['features'], train['labels']
```

```python
plt.imshow(xTrain[0])
```

```python
dictSigns = signnames.to_dict()['b']
```

```python
for signId in dictSigns:
    selectedSigns = xTrain[yTrain == signId]
    
    plt.figure(figsize = (15, 5))
    for i in range(9):
        plt.subplot(f'19{i + 1}')
        plt.imshow(selectedSigns[i])
        plt.axis('off')
    
plt.tight_layout()
plt.show()
```

```python
counter = Counter(yTrain).most_common()

labelsId, countLabels = zip(*counter)
ids = range(len(labelsId))

plt.figure(figsize = (15, 5))
plt.bar(ids, countLabels)
plt.xlabel('Znaki')

labels = [dictSigns[labelsId[x]] for x in labelsId]
plt.xticks(ids, labels, rotation = 'vertical')
plt.title('Znaki drogowe - ilo wystpie')
plt.show()
```
