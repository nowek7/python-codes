```python
def groupAndPlotOfCar(aFeatGroupBy, aFeatAgg = 'price_value', 
                      aAggFuncs = [np.mean, np.median, np.size], aSortFn = 'mean', 
                      aTop = 50, aHasSubPlots = True):
    return (
        carDf
          .groupby(aFeatGroupBy)[aFeatAgg]
          .agg(aAggFuncs)
          .sort_values(by = aSortFn, ascending = False)
          .head(aTop)
          ).plot(kind = 'bar', figsize = (15, 5), subplots = aHasSubPlots)
```

```python
groupAndPlotOfCar('param_marka-pojazdu');
```

```python
groupAndPlotOfCar('param_kraj-pochodzenia', aSortFn = 'size');
```

```python
groupAndPlotOfCar('param_kolor', aSortFn = 'mean');
```

```python
groupAndPlotOfCar('param_rok-produkcji', aSortFn = 'size');
```

```python
groupAndPlotOfCar('param_skrzynia-biegów', aSortFn = 'size');
```

```python
groupAndPlotOfCar('param_przebieg', aSortFn = 'mean');
```
